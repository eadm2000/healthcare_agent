```python
from transformers import pipeline

# Turn json_medical_conversation into a flat text string
conversation_text = ""
for entry in json_medical_conversation:
    speaker = f"Speaker {entry['speaker']}"
    for line in entry["dialogue"]:
        conversation_text += f"{speaker}: {line}\n"


summarizer = pipeline("summarization", model="facebook/bart-large-cnn")


conversation_chunk = conversation_text[:1024]  # simple truncation for this example

# Run summarization
summary = summarizer(conversation_chunk, max_length=150, min_length=40, do_sample=False)

print("📌 Summary:")
print(summary[0]['summary_text'])
```
